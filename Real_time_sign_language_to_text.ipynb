```python
import numpy as np
import mediapipe as mp
import cv2

# Initialiser MediaPipe Holistic
mp_holistic = mp.solutions.holistic
holistic = mp_holistic.Holistic(static_image_mode=False)

# Exemple de fonction pour traiter une vidÃ©o
def process_video(video_path):
    cap = cv2.VideoCapture(video_path)
    all = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Convertir l'image BGR en RGB
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Processer l'image avec MediaPipe Holistic
        results = holistic.process(frame_rgb)

        frame_x = []
        frame_y = []
        frame_z = []


        # Extraire les landmarks du visage
        if results.face_landmarks:
            for landmark in results.face_landmarks.landmark:
                frame_x.append(landmark.x)
                frame_y.append(landmark.y)
                frame_z.append(landmark.z)

        else :
                frame_x.extend(list(np.zeros(468)))
                frame_y.extend(list(np.zeros(468)))
                frame_z.extend(list(np.zeros(468)))

        # Extraire les landmarks de la main gauche
        if results.left_hand_landmarks:
            for landmark in results.left_hand_landmarks.landmark:
                frame_x.append(landmark.x)
                frame_y.append(landmark.y)
                frame_z.append(landmark.z)

        else :
                frame_x.extend(list(np.zeros(21)))
                frame_y.extend(list(np.zeros(21)))
                frame_z.extend(list(np.zeros(21)))
        # Extraire les landmarks de la main droite
        if results.right_hand_landmarks:
            for landmark in results.right_hand_landmarks.landmark:
                frame_x.append(landmark.x)
                frame_y.append(landmark.y)
                frame_z.append(landmark.z)

        else :
                frame_x.extend(list(np.zeros(21)))
                frame_y.extend(list(np.zeros(21)))
                frame_z.extend(list(np.zeros(21)))

        # Extraire les landmarks de la pose
        if results.pose_landmarks:
            for landmark in results.pose_landmarks.landmark:
                frame_x.append(landmark.x)
                frame_y.append(landmark.y)
                frame_z.append(landmark.z)

        else :
                frame_x.extend(list(np.zeros(33)))
                frame_y.extend(list(np.zeros(33)))
                frame_z.extend(list(np.zeros(33)))


        all.append(np.array([frame_x, frame_y, frame_z ]).reshape(len(frame_x), 3))


    cap.release()
    #holistic.close()

    # Convertir en tableaux NumPy
    all= np.array(all)


    return all

```

```python
# Utilisation de la fonction avec un chemin de vidÃ©o
video_path = "/content/drive/MyDrive/Tensorflow dev/real_time_sign_language/sign_video_data_dir/train/before/05734.mp4"
out = process_video(video_path)

print(f"out shape: {out.shape}")  # Devrait afficher (30, 542) pour une vidÃ©o de 30 frames
```
